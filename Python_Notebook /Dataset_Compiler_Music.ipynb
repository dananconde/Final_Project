```python
# Dependencies

import pandas as pd
import numpy as np
import os
```

```python
# Reading in datasets

reviews = pd.read_json("Resources/reviews_Musical_Instruments_5.json", lines=True)
reviews.head()
```

```python
# Filtering Columns

new_reviews = reviews[["asin", "overall", "reviewText", "summary"]]
new_reviews.columns = ["ASIN", "Star_Rating", "Review", "Summary"]
new_reviews.describe()
```

```python
def strip_punc_py(words):
  return "".join(char if (char.isalpha() or char == " ") else "" for char in words)

new_reviews["Clean_Review"] = new_reviews["Review"].map(strip_punc_py)

new_reviews.head(20)


```

```python
# Drop empty cells in 'Review' column
clean_reviews = new_reviews[["ASIN", "Star_Rating", "Summary", "Clean_Review"]]
clean_reviews = clean_reviews.dropna(subset=["Clean_Review"])
clean_reviews.Star_Rating = clean_reviews.Star_Rating.astype(int)
clean_reviews.head(20)
```

```python
# Exporting final training data as csv

clean_reviews.to_csv("musical_instruments_training.csv", index=False)
```
